### Problem Statement : Download Assignment files from iNeuron Dashboard manually,
### Automation - Rename the files in case any mismatch, Create directory from these file names with common names, and move these files to respective directory. Read all the files and and create a main document where write the file names as headings and content below.

```python
def FileHandling_operation():
    def rename_file(name):
        '''
        This function renames the file by removing '(' and ')' from the file name.
        The source file name has to be passed as an argument and the function will 
        rename the file in your local and return renamed file name.
        '''
        new_name = name.replace('(', '.').replace(')', '')  # Here we are replacing '(' with '.''
        os.rename(name, new_name)                           # Replacing the source file name with new file name
        return new_name   



    def folder(string):
        '''
        This function is to create folder name with the common strings in the files.
        It returns the string before any numeric character in it. Function also replaces '_' with '' 
        '''
        length = 0  
        for i in string:
            if i.isnumeric():
                length = string.find(i)                    # Finds the position where the numeric value is present in string
                break 
        return string[:length:].replace('_', '')           # returns the required string with out number and '_'

    import os
    import shutil
    from docx import Document


    newDoc = Document()                                    # Creating a new document to store the files 

    for i in os.listdir():                                 # Iterates over each file in the directory
        if i.find('.docx') >= 0:                           # Accepts only the .docx files

            #Rename file by replacing '(' &  ')'
            new_name = rename_file(i)

            # Getting the folder name from files
            folder_name = folder(new_name)

            # Creating a directory/folder from file name
            if not os.path.isdir(folder_name):
                os.mkdir(folder_name)

            # Adding the heading of the file name as level 2 in new document
            newDoc.add_heading(new_name, 2)

            # Creating an instance of the file we want to open 
            doc = Document(new_name)

            # Adding the content of the files to the new file
            for content in doc.paragraphs:
                newDoc.add_paragraph(content.text)

            # Adding a break to the file to start every file content in a new page
            newDoc.add_page_break()

            # Moving the files from current directory to new directory
            shutil.move(os.getcwd()+'\\'+new_name, os.getcwd()+'\\'+folder_name)


    # Saving the final document to a new file
    newDoc.save('MasterDocumet.docx')

    
FileHandling_operation()
```
